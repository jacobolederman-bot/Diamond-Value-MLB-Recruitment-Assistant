```python
#@title Upload your project files
import zipfile, os, shutil
from google.colab import files

#@markdown Check this box to re-upload a new zip
force_reupload = False  #@param {type:"boolean"}

if force_reupload:
    for f in os.listdir():
        if f.lower().endswith('.zip'):
            os.remove(f)
    if os.path.exists('hf_export'):
        shutil.rmtree('hf_export')

existing_zip = next((f for f in os.listdir() if f.lower().endswith('.zip')), None)
if existing_zip is None:
    print('⏳ While your files upload (large projects can take several minutes), start designing your app in Step 2 below.')
    print()
    uploaded = files.upload()
    if not uploaded:
        raise SystemExit('No file selected. Re-run this cell and pick project_export.zip when the file picker opens.')
    existing_zip = list(uploaded.keys())[0]

with zipfile.ZipFile(existing_zip, 'r') as z:
    z.extractall('hf_export')

extracted_files = [f for _, _, fs in os.walk('hf_export') for f in fs]
if not extracted_files:
    raise SystemExit('The zip extracted no files. Make sure you selected project_export.zip (not a different file).')

print(f'Extracted {existing_zip} -> hf_export/ ({len(extracted_files)} file(s))')
print()
print('Files in hf_export/:')
for root, _, fnames in sorted(os.walk('hf_export')):
    rel = os.path.relpath(root, 'hf_export')
    if rel == '.':
        for fname in sorted(fnames):
            print(f'  {fname}')
    else:
        n = len(fnames)
        print(f'  {rel}/ ({n} file{"s" if n != 1 else ""})')
```

Use the variables below for guidance!

**After brainstorming, fill in** your design choices in the first cell below. Then **run** the next cells to generate your prompt and package your project files. Your design answers become a prompt you'll paste into Replit's agent in Step 3.

<details><summary>What each field is for</summary>

| Variable | What to fill in |

```python
# ✏️ Brainstorm, then fill in your app details below. Make it yours!

import re, json, os
import ipywidgets as widgets
from IPython.display import display, Javascript, HTML

# App title - e.g. "X-Ray Diagnosis Assistant"
APP_NAME = "Diamond Value: MLB Recruitment Assistant"
PROJECT_DESCRIPTION = "A regression model trained on historical MLB player season statistics that predicts a player's market salary based on their offensive performance, helping a team decide which players are worth investing in based on their real value versus their cost."
# What your model does - e.g. "A CNN trained on chest X-ray images to classify pneumonia vs normal"
USER_ACTION = "the user can either describe what kind of player they're looking for, such as an under-the-radar breakout prospect worth betting on long-term versus a proven, established player who can contribute right away, and get matching recommendations, or search for a specific player by name, with the app suggesting similarly spelled names if there's no exact match; feel free to add other interaction modes or actions you think would make the app more useful or engaging"
# How results appear - e.g. "Prediction label + confidence bar"
RESULT_STYLE = "for player searches and recommendations, a scouting-report style verdict pairing predicted market value against actual salary with a headline call like 'Steal,' 'Fair Deal,' or 'Overpriced,' plus a visual comparison of value versus cost; for team analysis, a breakdown of the roster's statistical strengths and gaps paired with a recommended player profile to target, shown in that same scouting-report style so results feel consistent across the app"

# Visual style - e.g. "Medical, clean and minimal"
APP_THEME = "Sports analytics theme, clean and professional, styled like a baseball scouting dashboard"

# Optional extra features - e.g. "show sample images to try, fun facts panel"
EXTRA = "a team analysis mode that looks at a selected team's current roster and stats, then identifies gaps or weaknesses and recommends the type of player (based on statistical profile) that would best strengthen that team"
```

```python
#@title 🔧 Run this cell to generate your Replit prompt
#@markdown Click ▶ to run. This just formats your design choices clearly with instructions for Replit. Click **Show code** to see how it works exactly!

#@markdown *(Note: You could also write your own prompt, this just adds more instructions to make it run more reliably for Replit.)*

# ── Validation ──────────────────────────────────────────────────────────────
_required = {
    'APP_NAME': APP_NAME,
    'PROJECT_DESCRIPTION': PROJECT_DESCRIPTION,
    'USER_ACTION': USER_ACTION,
    'RESULT_STYLE': RESULT_STYLE,
    'APP_THEME': APP_THEME,
}
_missing = [k for k, v in _required.items() if not v.strip()]
if _missing:
    raise ValueError(f'Please go back and fill in: {", ".join(_missing)}')

# ── Build a slim, Replit-friendly prompt ────────────────────────────────────
# Keep it short. Let Replit's agent pick the stack and write good code.
# Only the bits it can't infer on its own (model_setup loading, Python required
# for inference) are mandatory.
_extra_line = f"\nEXTRA FEATURES: {EXTRA}" if EXTRA.strip() else ""

REPLIT_PROMPT = f"""Build a web app called "{APP_NAME}" from my project files. Unzip `replit_project.zip` into the project root first — all files live there.

WHAT IT DOES: {PROJECT_DESCRIPTION}
HOW USERS INTERACT: {USER_ACTION}
HOW RESULTS APPEAR: {RESULT_STYLE}
VISUAL STYLE: {APP_THEME}{_extra_line}

Frontend stack is up to you — Gradio, Streamlit, or React + a Python backend are all fine. Pick whichever makes the nicest app. A Python backend runs the model.

FIGURE OUT THE MODEL YOURSELF — don't assume, inspect the bundle:
- Load every file through `model_setup.paths["<filename>"]`, never by raw path. Read model_setup.py's docstring for loaders (joblib, Keras, Transformers).
- Use ALL provided files, not just the model — scalers, tokenizers, label maps, helper scripts, and data files are included because the app needs them.\n- If the bundle has multiple models or data sources for different features, build a section/mode for each — don't ship just one.
- Inspect the model and any data/schema files to determine the exact input shape and output type, then build the UI to match.

INPUTS: if the bundle includes sample model data (for example a sample_images folder, or a sample data file like a .csv / .json / .npz), give users preloaded examples they can try instantly (from their real sample data), matched to the input type. Use those real samples for the built-in examples, since they come from the model's own data and match what it expects. Do not substitute stock photos, web-search results, or AI-generated images for these built-in examples, which will not match the model and give meaningless predictions. This applies only to the preloaded examples: if there's no sample data in the bundle, skip them, and if the user later asks for a feature that needs outside images or data, that's fine. If the model takes many numeric features, let users pick or tweak an example instead of typing every field.
OUTPUTS: present results clearly and visually, matched to the output type — probability bars for classification, a value card for a single number, a ranked list for recommendations. No plain unstyled text.

RUNS ON REPLIT — must do:
- Bind the server to host 0.0.0.0 and the port Replit provides (the PORT env var if set) so the web preview loads.\n- Every process the app needs must start in PRODUCTION, not just development. If you split into multiple services (e.g. a Python backend), give each one a production run command, not only a development one, and confirm the published deployment launches them all. A backend with only a dev run config will fail when published.
- Install the included requirements.txt; if a pinned version conflicts with the environment, use the nearest compatible one.
- Some model files may download on first launch; show a loading state.
- CONFIGURE THE DEPLOYMENT CONFIG NOW, before finishing — not just the dev workflow. Set the deployment run command (and a build step if anything needs compiling) so the app is publishable immediately. Publishing uses the deployment config, NOT the dev Run button: if it is missing, clicking Publish fails with "Run command not found" / "Invalid run command".

VISUAL POLISH: Make it feel like a real product (based on the design choices above), not a default demo.

DONE WHEN: the app starts, the web preview loads, clicking a preloaded example returns a correct-looking result, AND the deployment config has a valid production run command so publishing works without a "Run command not found" error."""

# ── Save and display ────────────────────────────────────────────────────────
os.makedirs('hf_export', exist_ok=True)
with open('hf_export/.replit_prompt.txt', 'w') as f:
    f.write(REPLIT_PROMPT)

print('Design choices saved.\n')
print("Below is your Replit prompt. Click Copy, then paste into Replit in Step 3.")

display(HTML("<style>textarea { font-size: 14px !important; font-family: monospace; }</style>"))
prompt_box = widgets.Textarea(value=REPLIT_PROMPT, layout=widgets.Layout(width='100%', height='400px'))
copy_btn = widgets.Button(description='📋 Copy Replit prompt', button_style='primary',
                          layout=widgets.Layout(width='100%', height='40px'))
status = widgets.Output()
def copy(_):
    display(Javascript(f'navigator.clipboard.writeText({json.dumps(prompt_box.value)});'))
    with status:
        status.clear_output()
        print("Copied. Paste into Replit (you'll do this in Step 3).")
copy_btn.on_click(copy)
display(prompt_box, copy_btn, status)
```

> 🔒 **If your model files are large (over 150 MB),** the package cell below will ask for a HuggingFace Write token so it can offload them to HF Hub. Your app will download them on startup.

<details><summary>How to get a HuggingFace Write token (if asked)</summary>

1. Sign up at [huggingface.co/join](https://huggingface.co/join) and confirm your email.
2. From your profile photo, go to **Settings**, then **Access Tokens**, then click **+ Create new token**. Set the type to `Write` and copy it immediately.
3. Paste it when the cell asks.

</details>

```python
#@title 📦 Run this cell to package your project for Replit
#@markdown Click ▶ to run. This bundles your files into a `replit_project.zip` to load consistently, and hosts larger models on HuggingFace if needed (will ask for a HF token if so).

# Files at or above this size are offloaded to HuggingFace Hub so the Replit
# upload stays manageable. 100 MB keeps the zip well under Replit's 200 MB limit.
SIZE_THRESHOLD_MB = 150

import os, shutil, json, re, getpass

# Which extensions count as "model files" we might offload. Other files (.json,
# .csv, .txt, helpers.py) always stay in hf_export/ regardless of size.
HEAVY_EXTS = ('.h5', '.keras', '.pt', '.pth', '.bin', '.safetensors',
              '.joblib', '.pkl', '.onnx', '.tflite', '.gguf')

assert os.path.exists('hf_export'), "hf_export/ not found. Run the Upload step above first."

# ── Scan for heavy files ──────────────────────────────────────────────────────
threshold_bytes = SIZE_THRESHOLD_MB * 1024 * 1024
heavy_files = []  # list of (absolute path, path relative to hf_export/, size in bytes)
for root, _, fnames in os.walk('hf_export'):
    for fname in fnames:
        abs_path = os.path.join(root, fname)
        if fname.lower().endswith(HEAVY_EXTS) and os.path.getsize(abs_path) >= threshold_bytes:
            heavy_files.append((abs_path, os.path.relpath(abs_path, 'hf_export'), os.path.getsize(abs_path)))

total_mb = sum(os.path.getsize(os.path.join(r, f)) for r, _, fs in os.walk('hf_export') for f in fs) / (1024*1024)

REPO_ID = None  # set below only if we actually push

if not heavy_files:
    print(f"hf_export/ is {total_mb:.1f} MB. Nothing to offload.")
    print("(No HuggingFace token needed. All model files stay local in the zip.)")
else:
    print(f"hf_export/ is {total_mb:.1f} MB. {len(heavy_files)} file(s) over {SIZE_THRESHOLD_MB} MB will be offloaded to HuggingFace Hub:")
    for _, rel, sz in heavy_files:
        print(f"  • {rel}  ({sz / (1024*1024):.1f} MB)")
    print()

    # ── Get / install huggingface_hub ─────────────────────────────────────────
    try:
        from huggingface_hub import HfApi, create_repo
    except ImportError:
        import subprocess, sys
        subprocess.check_call([sys.executable, '-m', 'pip', 'install', '-q', 'huggingface_hub'])
        from huggingface_hub import HfApi, create_repo

    # ── Get HF credentials and project slug ───────────────────────────────────
    try:
        slug_source = APP_NAME  # noqa: F821. defined in earlier cell
    except NameError:
        slug_source = input("Project name (used to name your model repo): ").strip()
    PROJECT_SLUG = re.sub(r'[^a-z0-9-]', '', slug_source.lower().replace(' ', '-')).strip('-')
    assert PROJECT_SLUG, "Project name produced an empty slug. Use letters/numbers."

    HF_TOKEN    = getpass.getpass('Paste your HuggingFace Write token: ')
    HF_USERNAME = input('Your HuggingFace username: ').strip()
    REPO_ID = f"{HF_USERNAME}/{PROJECT_SLUG}-model"

    api = HfApi(token=HF_TOKEN)
    create_repo(repo_id=REPO_ID, repo_type='model', token=HF_TOKEN, exist_ok=True)
    print(f"\n✓ Model repo ready: huggingface.co/{REPO_ID}")

    # ── Upload heavy files ────────────────────────────────────────────────────
    for abs_path, rel_path, sz in heavy_files:
        print(f"  Uploading {rel_path} ({sz / (1024*1024):.1f} MB)...", end=' ', flush=True)
        api.upload_file(
            path_or_fileobj=abs_path,
            path_in_repo=rel_path,
            repo_id=REPO_ID,
            repo_type='model',
            commit_message=f'Upload {rel_path}',
        )
        print("✓")

    # ── Remove heavy files locally ────────────────────────────────────────────
    for abs_path, _, _ in heavy_files:
        os.remove(abs_path)
    print(f"\n✓ Removed {len(heavy_files)} heavy file(s) from hf_export/")
    print(f"✓ Your model files live at: https://huggingface.co/{REPO_ID}")

# ── ALWAYS write model_config.json and model_setup.py ────────────────────────
# Even for small projects, this gives every app.py a single consistent loading
# pattern: `import model_setup; ...load(model_setup.paths["model.h5"])`.
# When REPO_ID is None, model_setup just returns local paths.
hub_files = [rel for _, rel, _ in heavy_files] if heavy_files else []
model_config = {
    'hub_repo_id': REPO_ID,        # None if all files local
    'hub_files': hub_files,        # files that live on HF Hub
}
with open('hf_export/model_config.json', 'w') as f:
    json.dump(model_config, f, indent=2)

model_setup_code = '''"""Auto-generated. Provides `paths`. a dict mapping each project file to a
local path your app.py can load. Files that are too big for Replit live on the
HuggingFace Hub and are downloaded on first use; everything else stays local.

Usage in app.py:

    import model_setup

    # For Keras / TensorFlow:
    import tensorflow as tf
    model = tf.keras.models.load_model(model_setup.paths["my_model.h5"])

    # For scikit-learn / joblib:
    from joblib import load
    model = load(model_setup.paths["my_model.joblib"])

    # For HuggingFace Transformers (whole directory):
    from transformers import AutoModelForSequenceClassification
    model = AutoModelForSequenceClassification.from_pretrained(model_setup.dir)

`paths` works for any filename present locally OR uploaded to HF Hub. Downloaded
Hub files are cached, so they only download once per environment.
"""
import json, os

_here = os.path.dirname(os.path.abspath(__file__))
dir = _here  # exposed for transformers/from_pretrained patterns

# 1. Every local file is reachable by its filename (including subdirectories)
paths = {}
for _root, _dirs, _fnames in os.walk(_here):
    for _fname in _fnames:
        _full = os.path.join(_root, _fname)
        _rel  = os.path.relpath(_full, _here)
        paths[_rel]   = _full
        paths[_fname] = _full

# 2. If model_config.json exists and points to HF Hub files, download them
#    and override the local entry. If config is missing or empty, we just use
#    local files. safe fallback for projects that don't need HF Hub.
_cfg_path = os.path.join(_here, "model_config.json")
if os.path.exists(_cfg_path):
    with open(_cfg_path) as _f:
        _cfg = json.load(_f)
    if _cfg.get("hub_repo_id"):
        from huggingface_hub import hf_hub_download
        for _fname in _cfg.get("hub_files", []):
            paths[_fname] = hf_hub_download(repo_id=_cfg["hub_repo_id"], filename=_fname)
'''
with open('hf_export/model_setup.py', 'w') as f:
    f.write(model_setup_code)
print(f"\n✓ Wrote hf_export/model_setup.py (your app.py uses this to load files)")

# ── Clean up requirements.txt for Replit ────────────────────────────────────
# 1. Strip the `spaces` package (HF-Spaces-only, breaks Replit installs)
# 2. Ensure huggingface_hub is present (needed at runtime for HF Hub downloads)
req_path = 'hf_export/requirements.txt'
existing_reqs = ''
if os.path.exists(req_path):
    with open(req_path) as f:
        existing_reqs = f.read()
else:
    print('⚠️  No requirements.txt found in hf_export/. A minimal one will be created.')
    print('   If your app fails to run in Replit, ask it:')
    print('   "Create a requirements.txt listing every package app.py imports."')
    print()

cleaned_lines = []
for line in existing_reqs.splitlines():
    stripped = line.strip()
    if stripped.lower().split('=')[0].split('<')[0].split('>')[0].strip() == 'spaces':
        continue
    cleaned_lines.append(line)

if not any('huggingface_hub' in l.lower() for l in cleaned_lines):
    cleaned_lines.append('huggingface_hub')

with open(req_path, 'w') as f:
    f.write('\n'.join(cleaned_lines).rstrip() + '\n')

# ── Zip up the project for Replit ───────────────────────────────────────────
SLIM_ZIP = 'replit_project.zip'
if os.path.exists(SLIM_ZIP):
    os.remove(SLIM_ZIP)
shutil.make_archive('replit_project', 'zip', 'hf_export')
final_size_mb = os.path.getsize(SLIM_ZIP) / (1024*1024)
print(f"\n{SLIM_ZIP} is {final_size_mb:.1f} MB.")

# Warn if still too big for Replit
if final_size_mb > 180:
    print(f"\n⚠️  WARNING: {SLIM_ZIP} is too big for Replit's 200 MB upload limit.")
    print("   Options:")
    print(f"   1. Re-run this cell with SIZE_THRESHOLD_MB = 20 (or 10) to push smaller files to HF Hub.")
    print("   2. If you have a large corpus/ folder, reduce the number of files.")
    print("   3. Identify the biggest remaining files:")
    sizes = sorted(
        ((os.path.getsize(os.path.join(r, f)) / (1024*1024), os.path.relpath(os.path.join(r, f), 'hf_export'))
         for r, _, fs in os.walk('hf_export') for f in fs),
        reverse=True,
    )[:5]
    for sz, name in sizes:
        print(f"      • {name}  ({sz:.1f} MB)")

# ── Auto-download to student's Downloads folder so Step 3 just works ─────────
try:
    from google.colab import files
    files.download(SLIM_ZIP)
    print(f"\n✓ Downloaded {SLIM_ZIP} to your computer's Downloads folder.")
    print(f"  Drag THIS file (not the original project_export.zip) into Replit in Step 3.")
except ImportError:
    print(f"\n(Outside Colab. To download {SLIM_ZIP}:")
    print("  Click the folder icon in Colab's left sidebar,")
    print(f" right-click {SLIM_ZIP}, and choose Download.)")
```
